```python
dataset['year']=(dataset['lowest_yearly_earnings']+dataset['highest_yearly_earnings'])
print(dataset['year'].describe())
```

```python
def outl(data):
  q1=data.quantile(0.25)
  q3=data.quantile(0.75)
  iqr=q3-q1
  upp=q3+1.5*iqr
  low=q1-1.5*iqr
  return low,upp
print(outl(dataset['year']))
```

All values below low value and above upp value are outliers

```python
sns.boxplot(y='year',data=dataset)
plt.title('Yearly earnings')
plt.show()
```
